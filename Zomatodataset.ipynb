```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings
```

# observations
1. When Ratings is between 4.5 to 4.9 --> Excellent
2. When Ratings is between 3.5 to 3.9 --> Good
3. When Ratings is between 2.5 to 2.9 --> Average
4. When Ratings is between 1.8 to 2.4 --> Poor
5. There are around 2148 people who have not given any ratings

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x='Aggregate rating',y='Rating Count',data=Ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=Ratings, palette=['Blue','red','orange','yellow','green','darkgreen'])
```

Observations
1.Not Rated value is too high
2.max number of rating is between 2.5 to 3.4
